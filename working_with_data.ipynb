```python
import numpy as np
np.sum(df['total_vaccinations'].isna())
for col in df.columns:
    perc = np.mean(df[col].isna())
    print(f'{col}: {round(perc*100)} %')
```

```python
## eliminating the not a number lines
df.dropna().reset_index(drop=True)
```

```python
df.describe(
)
```

```python
frequency = df.median().values
frequency
```

```python
cols = ['total_vaccinations',
       'people_vaccinated', 
        'people_fully_vaccinated',
       'daily_vaccinations_raw',
        'daily_vaccinations',
       'total_vaccinations_per_hundred',
        'people_vaccinated_per_hundred',
       'people_fully_vaccinated_per_hundred', 
        'daily_vaccinations_per_million']
```

```python
for i in range(len(frequency)):
    df[cols[i]] = df[cols[i]].fillna(frequency[i])

missingno.matrix(df)
```
