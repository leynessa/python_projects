Crie uma função que recebe como variável de entrada uma lista e retorna outra lista contendo somente elementos do tipo *string*. 

````python

def choose_string(lst):
  new_list = []
  for i in lst:
    if type(i) == str:
      new_list.append(i)  
  return new_list 

lst="1", 2,3,"8"
````
